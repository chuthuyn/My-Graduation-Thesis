## Remove Outliers

```python
#convert outliers to null values

for x in ['trestbps']:
    q75,q25 = np.percentile(patient_df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    patient_df.loc[patient_df[x] < min,x] = np.nan
    patient_df.loc[patient_df[x] > max,x] = np.nan
    
for x in ['chol']:
    q75,q25 = np.percentile(patient_df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    patient_df.loc[patient_df[x] < min,x] = np.nan
    patient_df.loc[patient_df[x] > max,x] = np.nan

for x in ['thalach']:
    q75,q25 = np.percentile(patient_df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    patient_df.loc[patient_df[x] < min,x] = np.nan
    patient_df.loc[patient_df[x] > max,x] = np.nan
```

```python
#count all null values had been converted

tt = pd.isna(patient_df['trestbps']).sum()
print('null values of tresbps:',tt)
tc = pd.isna(patient_df['chol']).sum()
print('null values of cholesterol:',tc)
t2 = pd.isna(patient_df['thalach']).sum()
print('null values of thalach:',t2)
```

## Logistic Regression

```python
import statsmodels.api as sm
from statsmodels.tools import add_constant as add_constant

patient_df_constant = add_constant(patient_df)
cols = patient_df_constant.columns[:-1]
patient_df_constant.head()
logit_model=sm.Logit(patient_df['target'], patient_df_constant[cols])
result=logit_model.fit()
print(result.summary())
```

```python
def back_feature_elem (data_frame,dep_var,col_list):
    while len(col_list)>0 :
        model=sm.Logit(dep_var,data_frame[col_list])
        result=model.fit(disp=0)
        largest_pvalue=round(result.pvalues,3).nlargest(1)
        if largest_pvalue[0]<(0.05):
            return result
            break
        else:
            col_list=col_list.drop(largest_pvalue.index)


result=back_feature_elem(patient_df_constant,patient_df.target,cols)
print(result.summary())
```

## Split data and training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn.metrics import confusion_matrix, accuracy_score
new_features = patient_df.drop(columns = ['age', 'fbs'])
X = new_features.drop(columns = 'target')
y = new_features['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 42)
log_model = LogisticRegression(solver = 'liblinear').fit(X_train, y_train)
y_pred = log_model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print("Logistic Regression Model's Accuracy:", acc*100, '%')
```
